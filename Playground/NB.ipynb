```python
import json
from pprint import pprint
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fileName = '/home/jsteffen/Jason06.txt'
with open(fileName) as f:
    d = json.load(f)
    f.close()
    
```

```python
#comp = 7 # 11    
TrialList = [7, 11, 15, 19, 23, 27, 31, 35, 39, 43]
count = 1
for comp in TrialList:
    e = d[comp + 1]
    f = e['webgazer_data']
    x = []
    y = []
    t = []
    for i in f:
        x.append(i['x'])
        y.append(i['y'])
        t.append(i['t'])

    Dt = [0, np.diff(t)]
    data = {'x':x,'y':y,'t':t}
    
    df = pd.DataFrame(data)    
    df['Dt'] = df['t'].diff()

    plt.plot(x,y)
    plt.scatter(df['x'],df['y'],s=df['Dt']/2,alpha = 0.2)
    plt.xlim(00, 1900)
    plt.ylim(00, 800)
    ax = plt.gca()
    ax.set_aspect('equal', adjustable='box')

    Target1 = e['webgazer_targets']['#TrackingTarget_0']
    Target2 = e['webgazer_targets']['#TrackingTarget_2']
    Target3 = e['webgazer_targets']['#TrackingTarget_6']
    Target4 = e['webgazer_targets']['#TrackingTarget_8']
    plt.plot(Target1['x'],Target1['y'],'rx')
    plt.plot(Target2['x'],Target2['y'],'rx')
    plt.plot(Target3['x'],Target3['y'],'rx')
    plt.plot(Target4['x'],Target4['y'],'rx')
    plt.plot(df['x'][0], df['y'][0], 'g>')
    plt.plot(df['x'].iloc[-1], df['y'].iloc[-1], 'rs')
    plt.title("Trial: "+ str(count))
    plt.show()
    count += 1

```
